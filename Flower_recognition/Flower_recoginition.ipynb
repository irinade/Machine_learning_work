# EfficientNet pretrain model


```python
%env SM_FRAMEWORK=tf.keras
import efficientnet.keras as efn
from tensorflow import keras
```

```python
# Add our data-augmentation parameters to ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255., rotation_range = 40, width_shift_range = 0.2, height_shift_range = 0.2, shear_range = 0.2, zoom_range = 0.2, horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1.0/255.)

train_generator = train_datagen.flow_from_directory(train_dir, batch_size = 20, target_size = (224, 224))

validation_generator = test_datagen.flow_from_directory( validation_dir, batch_size = 20, target_size = (224, 224))
```

```python
base_model = efn.EfficientNetB0(input_shape = (224, 224, 3), include_top = False, weights = 'imagenet')
```

```python
for layer in base_model.layers:
    layer.trainable = False
```

```python
#%tensorflow_version 1.x
```

```python
#from tensorflow.keras.layers import Dense, Flatten, Dropout

from keras.layers import Dense, Flatten, Dropout

x = base_model.output
x = Flatten()(x)
x = Dense(1024, activation="relu")(x)
x = Dropout(0.5)(x)

# Add a final sigmoid layer with 1 node for classification output
predictions = Dense(5, activation="sigmoid")(x)


model_final = Model(base_model.input, predictions)
```

```python
predictions.shape
```

```python
from tensorflow.keras.optimizers import RMSprop

opt = RMSprop(lr=0.0001, decay=1e-6)

model_final.compile(optimizer = opt,loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
hist = model_final.fit_generator(train_generator, validation_data = validation_generator, steps_per_epoch = 50, epochs = 10)
```

```python
import matplotlib.pyplot as plt


def plot_hist(hist):
    plt.plot(hist.history["accuracy"])
    plt.plot(hist.history["val_accuracy"])
    plt.title("model accuracy")
    plt.ylabel("accuracy")
    plt.xlabel("epoch")
    plt.legend(["train", "validation"], loc="upper left")
    plt.show()


plot_hist(hist)
```
